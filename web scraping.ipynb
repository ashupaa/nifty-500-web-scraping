# WEB SCRAPING USING PYTHON LIBRARIES

```python
from bs4 import BeautifulSoup
```

BeautifulSoup is a Python library used for web scraping and parsing HTML and XML documents. It helps to extract data from web pages by navigating and searching the parse tree easily.

```python
import requests
```

request library is used to for handling responses

```python
import pandas as pd
```

 Pandas library is used to analyze, clean, store data and manipulate data

```python
url = " https://en.wikipedia.org/wiki/NIFTY_500"
```

```python
page = requests.get(url)
page
```

requests.get(url) fetches the webpage.

```python
soup = BeautifulSoup(page.text , 'html')
```

```python
print(soup.prettify())
```

```python
table = soup.find_all('table')[2]
```

```python
table_col_head = table.find_all('tr')[0]
```

 code_head will extract me the colmuns heading in form of a list

```python
col_head =[head.text.strip() for head in table_col_head]
col_head
```

 below code will extract all the data in each rows 

```python
filtered_col_head = [item for item in col_head if item!= ""]
print(filtered_col_head)
```

```python
df=pd.DataFrame(columns = filtered_col_head)
df
```

```python
row_data=table.find_all('tr')
```

```python
for row in row_data[1:]:
    row_val = row.find_all('td')
    each_row = [ row_inside.text.strip() for row_inside in row_val]
    print(each_row)

    length = len(df)
    df.loc[length]= each_row
```
